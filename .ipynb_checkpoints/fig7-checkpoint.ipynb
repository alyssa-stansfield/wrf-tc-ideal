```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
f298 = xr.open_dataset('./data_files/storm_mode_frequencies_298K_alltimes.nc')
DWCC_298 = f298.DWCC_freq
BSR_298 = f298.BSR_freq
f298.close()
```

```python
f300 = xr.open_dataset('./data_files/storm_mode_frequencies_300K_alltimes.nc')
DWCC_300 = f300.DWCC_freq
BSR_300 = f300.BSR_freq
f300.close()
```

```python
f302 = xr.open_dataset('./data_files/storm_mode_frequencies_302K_alltimes.nc')
DWCC_302 = f302.DWCC_freq
BSR_302 = f302.BSR_freq
f302.close()
```

```python
f304 = xr.open_dataset('./data_files/storm_mode_frequencies_304K_alltimes.nc')
DWCC_304 = f304.DWCC_freq
BSR_304 = f304.BSR_freq
f304.close()
```

```python
fig7, ax = plt.subplots(3,2,figsize=(10,15))


dwcc_levs = np.arange(-100,108,8)
cmap_dwcc = mcolors.LinearSegmentedColormap.from_list(name='red_white_blue', 
                                                 colors =[(0, 0, 1), 
                                                          (1, 1., 1), 
                                                          (1, 0, 0)],
                                                 N=len(dwcc_levs)-1,
                                                 )

occ_levs = np.arange(-4.6,5.0,0.4)
cmap_occ = mcolors.LinearSegmentedColormap.from_list(name='red_white_blue', 
                                                 colors =[(0, 0, 1), 
                                                          (1, 1., 1), 
                                                          (1, 0, 0)],
                                                 N=len(occ_levs)-1)

dcc_levs = np.arange(-0.9,1.02,0.12)
cmap_dcc = mcolors.LinearSegmentedColormap.from_list(name='red_white_blue', 
                                                 colors =[(0, 0, 1), 
                                                          (1, 1., 1), 
                                                          (1, 0, 0)],
                                                 N=len(dcc_levs)-1,
                                                 )

wcc_levs = np.arange(-7.6,8.4,0.8)
cmap_wcc = mcolors.LinearSegmentedColormap.from_list(name='red_white_blue', 
                                                 colors =[(0, 0, 1), 
                                                          (1, 1., 1), 
                                                          (1, 0, 0)],
                                                 N=len(wcc_levs)-1,
                                                 )

dwcc = ax[0,0].contourf(DWCC_300-DWCC_298,levels=dwcc_levs,cmap=cmap_dwcc)
ax[0,0].set_xlim(100,400)
ax[0,0].set_xticks([])
ax[0,0].set_ylim(100,400)
ax[0,0].set_title("DWCC",fontsize=26)
ax[0,0].set_ylabel("300 K - 298 K",fontsize=22)
diff_DWCC_300_298 = DWCC_300.to_numpy()-DWCC_298.to_numpy()
diff_DWCC_300_298[diff_DWCC_300_298==0] = np.nan
ax[0,0].text(330,375,"%.2f" % round(np.nanmean(diff_DWCC_300_298), 2), fontsize=18)

bsr = ax[0,1].contourf(BSR_300-BSR_298,levels=dwcc_levs,cmap=cmap_dwcc)
ax[0,1].set_xlim(100,400)
ax[0,1].set_xticks([])
ax[0,1].set_yticks([])
ax[0,1].set_ylim(100,400)
ax[0,1].set_title("BSR",fontsize=26)
diff_BSR_300_298 = BSR_300.to_numpy()-BSR_298.to_numpy()
diff_BSR_300_298[diff_BSR_300_298==0] = np.nan
ax[0,1].text(330,375,"%.2f" % round(np.nanmean(diff_BSR_300_298), 2), fontsize=18)

ax[1,0].contourf(DWCC_302-DWCC_300,levels=dwcc_levs,cmap=cmap_dwcc)
ax[1,0].set_xlim(100,400)
ax[1,0].set_ylim(100,400)
ax[1,0].set_ylabel("302 K - 300 K",fontsize=22)
diff_DWCC_302_300 = DWCC_302.to_numpy()-DWCC_300.to_numpy()
diff_DWCC_302_300[diff_DWCC_302_300==0] = np.nan
ax[1,0].text(330,375,"%.2f" % round(np.nanmean(diff_DWCC_302_300), 2), fontsize=18)

ax[1,1].contourf(BSR_302-BSR_300,levels=dwcc_levs,cmap=cmap_dwcc)
ax[1,1].set_xlim(100,400)
ax[1,1].set_ylim(100,400)
ax[1,1].set_yticks([])
diff_BSR_302_300 = BSR_302.to_numpy()-BSR_300.to_numpy()
diff_BSR_302_300[diff_BSR_302_300==0] = np.nan
ax[1,1].text(330,375,"%.2f" % round(np.nanmean(diff_BSR_302_300), 2), fontsize=18)

cax1 = plt.axes((0.125, 0.065, 0.77, 0.03))
cbar1 = plt.colorbar(dwcc, cax=cax1, orientation="horizontal",spacing='uniform')
cbar1.ax.tick_params(labelsize=20)
cbar1.ax.set_xticks(ticks=[-100,-68, -36, 0, 36, 68, 100], labels=['-100','-68','-36','0','36','68','100'])
cbar1.set_label(label='%',weight='bold',size=30)

ax[0,0].text(110,375,"a",fontsize=22,weight="bold")
ax[1,0].text(110,375,"c",fontsize=22,weight="bold")
ax[0,1].text(110,375,"b",fontsize=22,weight="bold")
ax[1,1].text(110,375,"d",fontsize=22,weight="bold")
ax[2,0].text(110,375,"e",fontsize=22,weight="bold")
ax[2,1].text(110,375,"f",fontsize=22,weight="bold")

dwcc = ax[2,0].contourf(DWCC_304-DWCC_302,levels=dwcc_levs,cmap=cmap_dwcc)
ax[2,0].set_xlim(100,400)
ax[2,0].set_ylim(100,400)
ax[2,0].set_ylabel("304 K - 302 K",fontsize=22)
diff_DWCC_304_302 = DWCC_304.to_numpy()-DWCC_302.to_numpy()
diff_DWCC_304_302[diff_DWCC_304_302==0] = np.nan
ax[2,0].text(340,375,"%.2f" % round(np.nanmean(diff_DWCC_304_302), 2), fontsize=18)

bsr = ax[2,1].contourf(BSR_304-BSR_302,levels=dwcc_levs,cmap=cmap_dwcc)
ax[2,1].set_xlim(100,400)
ax[2,1].set_ylim(100,400)
diff_BSR_304_302 = BSR_304.to_numpy()-BSR_302.to_numpy()
diff_BSR_304_302[diff_BSR_304_302==0] = np.nan
ax[2,1].text(340,375,"%.2f" % round(np.nanmean(diff_BSR_304_302), 2), fontsize=18)
```
