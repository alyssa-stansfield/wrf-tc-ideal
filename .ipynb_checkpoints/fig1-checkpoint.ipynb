```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#temperature, moisture, and height profiles from default idealized TC test case
f1 = pd.read_csv("./data_files/default_sounding.csv")
theta_default = f1["T"].drop(index=0) #K
z_default = f1["P"].drop(index=0)/1000. #convert from m to km
q_default = f1["Q"].drop(index=0) #g/kg
```

```python
#temperature, moisture, and height profiles from last 30 days of 298 K RCE simulation
f2 = pd.read_csv("./data_files/RCE_298K_init_profs.csv",names=["Z","T","Q","U","V"],sep="\t")
theta_298 = f2["T"].drop(index=0) #K
z_298 = f2["Z"].drop(index=0)/1000. #convert to km
q_298 = f2["Q"].drop(index=0) #g/kg
```

```python
#temperature, moisture, and height profiles from last 30 days of 300 K RCE simulation
f3 = pd.read_csv("./data_files/RCE_300K_init_profs.csv",names=["Z","T","Q","U","V"],sep="\t")
theta_300 = f3["T"].drop(index=0) #K
z_300 = f3["Z"].drop(index=0)/1000. #convert to km
q_300 = f3["Q"].drop(index=0) #g/kg
```

```python
#temperature, moisture, and height profiles from last 30 days of 302 K RCE simulation
f4 = pd.read_csv("./data_files/RCE_302K_init_profs.csv",names=["Z","T","Q","U","V"],sep="\t")
theta_302 = f4["T"].drop(index=0) #K
z_302 = f4["Z"].drop(index=0)/1000. #convert to km
q_302 = f4["Q"].drop(index=0) #g/kg
```

```python
#temperature, moisture, and height profiles from last 30 days of 304 K RCE simulation
f5 = pd.read_csv("./data_files/RCE_304K_init_profs.csv",names=["Z","T","Q","U","V"],sep="\t")
theta_304 = f5["T"].drop(index=0) #K
z_304 = f5["Z"].drop(index=0)/1000. #convert to km
q_304 = f5["Q"].drop(index=0) #g/kg
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.plot(theta_default,z_default,'k--',label="Default")
ax1.plot(theta_298,z_298,c="gold",label="298 K")
ax1.plot(theta_300,z_300,c="darkorange",label="300 K")
ax1.plot(theta_302,z_302,'r',label="302 K")
ax1.plot(theta_304,z_304,c="maroon",label="304 K")

ax2.plot(q_default,z_default,'k--')
ax2.plot(q_298,z_298,c="gold")
ax2.plot(q_300,z_300,c="darkorange")
ax2.plot(q_302,z_302,'r')
ax2.plot(q_304,z_304,c="maroon")

ax1.set_ylim(0.5,15.5)
ax2.set_ylim(0.5,15.5)

ax1.set_xlim(296,400)
ax1.set_ylabel("Height [km]",fontsize=20)
ax1.set_xlabel("Potential Temperature [K]",fontsize=16)
ax2.set_xlabel("Specific Humidity [g/kg]",fontsize=16)
ax1.legend(fontsize=11,loc=4)
ax2.set_xlim(0,22)

ax1.text(302,14,"a",fontsize=20,weight="bold")
ax2.text(1.6,14,"b",fontsize=20,weight="bold")
```
