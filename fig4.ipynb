```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_298 = "./data_files/radial_vertical_profiles_298K.nc"
#only read in values starting 24 hours after spin-up
f_298 = xr.open_dataset(file_298, engine='netcdf4')
r_298 = f_298.r #distance away from TC center for x-axis
z_298 = f_298.z_mean[95:,:] #vertical distance from surface for y-axis
W_298 = f_298.W[95:,:,:] #upward wind speed
Vrad_298 = f_298.v_rad[95:,:,:] #radial wind speed
f_298.close()
```

```python
file_300 = "./data_files/radial_vertical_profiles_300K.nc"

f_300 = xr.open_dataset(file_300, engine='netcdf4')
r_300 = f_300.r
z_300 = f_300.z_mean[95:,:]
W_300 = f_300.W[95:,:,:]
Vrad_300 = f_300.v_rad[95:,:,:]
f_300.close()
```

```python
file_302 = "./data_files/radial_vertical_profiles_302K.nc"

f_302 = xr.open_dataset(file_302, engine='netcdf4')
r_302 = f_302.r
z_302 = f_302.z_mean[95:,:]
W_302 = f_302.W[95:,:,:]
Vrad_302 = f_302.v_rad[95:,:,:]
f_302.close()
```

```python
file_304 = "./data_files/radial_vertical_profiles_304K.nc"

f_304 = xr.open_dataset(file_304, engine='netcdf4')
r_304 = f_304.r
z_304 = f_304.z_mean[95:,:]
W_304 = f_304.W[95:,:,:]
Vrad_304 = f_304.v_rad[95:,:,:]
f_304.close()
```

```python
fig4, ax = plt.subplots(4,1,figsize=(7,12))

ax[0].contourf(r_298,np.nanmean(z_298,axis=0),np.nanmean(W_298,axis=0),levels=[-0.25,0,0.25,0.5,0.75,1,1.25,1.5,1.75,2,2.25,2.5,2.75])
ax[0].contour(r_298,np.nanmean(z_298,axis=0),np.nanmean(Vrad_298,axis=0),colors='w',levels=[-10,-5,-2.5,2.5,5,10])
ax[0].set_ylim(0.2,17)
ax[0].set_xlim(2,400)
ax[0].set_xticks([])
ax[0].set_ylabel("Height [km]",fontsize=20)
ax[0].set_title("298 K", fontsize=24)

ax[1].contourf(r_300,np.nanmean(z_300,axis=0),np.nanmean(W_300,axis=0),levels=[-0.25,0,0.25,0.5,0.75,1,1.25,1.5,1.75,2,2.25,2.5,2.75])
ax[1].contour(r_300,np.nanmean(z_300,axis=0),np.nanmean(Vrad_300,axis=0),colors='w',levels=[-10,-5,-2.5,2.5,5,10])
ax[1].set_ylim(0.2,17)
ax[1].set_xlim(2,400)
ax[1].set_xticks([])
ax[1].set_ylabel("Height [km]",fontsize=20)
ax[1].set_title("300 K", fontsize=24)

ax[2].contourf(r_302,np.nanmean(z_302,axis=0),np.nanmean(W_302,axis=0),levels=[-0.25,0,0.25,0.5,0.75,1,1.25,1.5,1.75,2,2.25,2.5,2.75])
ax[2].contour(r_302,np.nanmean(z_302,axis=0),np.nanmean(Vrad_302,axis=0),colors='w',levels=[-10,-5,-2.5,2.5,5,10])
ax[2].set_ylim(0.2,17)
ax[2].set_xlim(2,400)
ax[2].set_xticks([])
ax[2].set_ylabel("Height [km]",fontsize=20)
ax[2].set_title("302 K", fontsize=24)

c = ax[3].contourf(r_304,np.nanmean(z_304,axis=0),np.nanmean(W_304,axis=0),levels=[-0.25,0,0.25,0.5,0.75,1,1.25,1.5,1.75,2,2.25,2.5,2.75])
ax[3].contour(r_304,np.nanmean(z_304,axis=0),np.nanmean(Vrad_304,axis=0),colors='w',levels=[-10,-5,-2.5,2.5,5,10])
ax[3].set_ylim(0.2,17)
ax[3].set_xlim(2,400)
ax[3].set_xlabel("Distance from TC center [km]",fontsize=20)
ax[3].set_ylabel("Height [km]",fontsize=20)
ax[3].set_title("304 K", fontsize=24)

ax[0].text(20,15,"a",fontsize=16,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[1].text(20,15,"b",fontsize=16,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[2].text(20,15,"c",fontsize=16,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[3].text(20,15,"d",fontsize=16,weight="bold",backgroundcolor="white",ha="center", va="center")


cax1 = plt.axes((0.125, 0.045, 0.77, 0.02))
cbar1 = plt.colorbar(c, cax=cax1, orientation="horizontal",spacing='uniform')
cbar1.ax.tick_params(labelsize=16)
cbar1.set_label(label='m/s',weight='bold',size=18)
```
