```python
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
import pandas as pd
```

```python
f_298 = xr.open_dataset("./data_files/microphysics_means_298K_alltimes.nc")
Qr298 = f_298.Qr[:,:,:].values
Qs298 = f_298.Qs[:,:,:].values
Qc298 = f_298.Qc[:,:,:].values
Qi298 = f_298.Qi[:,:,:].values
Qg298 = f_298.Qg[:,:,:].values
r298 = f_298.r
z298 = f_298.lev
time298 = f_298.time
f_298.close()

Qr298[Qr298<=0] = np.nan
Qs298[Qs298<=0] = np.nan
Qc298[Qc298<=0] = np.nan
Qi298[Qi298<=0] = np.nan
Qg298[Qg298<=0] = np.nan

ts298 = pd.read_csv("./data_files/TC_outer_inner_evolution_298K.csv")
r_inner_298 = ts298.rinner_v[95:]
r_outer_298 = ts298.router_v[95:]
r_inner_mean_298 = np.nanmean(r_inner_298)
r_outer_mean_298 = np.nanmean(r_outer_298)
```
