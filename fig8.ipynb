```python
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
```

```python
f_298 = xr.open_dataset("./data_files/microphysics_sums_298K_alltimes.nc")
Qr298 = f_298.Qr[:,:,:].values #rain mixing ratio
Qs298 = f_298.Qs[:,:,:].values #snow mixing ratio
Qc298 = f_298.Qc[:,:,:].values #cloud liquid mixing ratio
Qi298 = f_298.Qi[:,:,:].values #cloud ice mixing ratio
Qg298 = f_298.Qg[:,:,:].values #graupel mixing ratio
r298 = f_298.r #distance from TC center for x-axis
z298 = f_298.lev #height for y-axis
time298 = f_298.time
f_298.close()

Qr298[Qr298<=0] = np.nan
Qs298[Qs298<=0] = np.nan
Qc298[Qc298<=0] = np.nan
Qi298[Qi298<=0] = np.nan
Qg298[Qg298<=0] = np.nan
```

```python
f_300 = xr.open_dataset("./data_files/microphysics_sums_300K_alltimes.nc")
Qr300 = f_300.Qr[:,:,:].values
Qs300 = f_300.Qs[:,:,:].values
Qc300 = f_300.Qc[:,:,:].values
Qi300 = f_300.Qi[:,:,:].values
Qg300 = f_300.Qg[:,:,:].values
r300 = f_300.r
z300 = f_300.lev
f_300.close()

Qr300[Qr300<=0] = np.nan
Qs300[Qs300<=0] = np.nan
Qc300[Qc300<=0] = np.nan
Qi300[Qi300<=0] = np.nan
Qg300[Qg300<=0] = np.nan
```

```python
f_304 = xr.open_dataset("./data_files/microphysics_sums_304K_alltimes.nc")
Qr304 = f_304.Qr[:,:,:].values
Qs304 = f_304.Qs[:,:,:].values
Qc304 = f_304.Qc[:,:,:].values
Qi304 = f_304.Qi[:,:,:].values
Qg304 = f_304.Qg[:,:,:].values
r304 = f_304.r
z304 = f_304.lev
f_304.close()

Qr304[Qr304<=0] = np.nan
Qs304[Qs304<=0] = np.nan
Qc304[Qc304<=0] = np.nan
Qi304[Qi304<=0] = np.nan
Qg304[Qg304<=0] = np.nan
```

```python
fig8, ax = plt.subplots(2,4,figsize=(30,20))

Qr = ax[0,0].contourf(r298,z298,np.nanmean(Qr298,axis=0),cmap="PuBuGn",levels=np.arange(50,1200,100))
ax[0,0].contour(r298,z298,np.nanmean(Qs298,axis=0),cmap="cool",levels=(np.arange(500,2500,250)))
ax[0,0].contour(r298,z298,np.nanmean(Qg298,axis=0),colors="k",levels=(np.arange(50,250,50)))
ax[0,0].set_ylim(1,18)
ax[0,0].set_xlim(0,500)
ax[0,0].set_ylabel("Height [km]",fontsize=34)
cax = plt.axes((0.925, 0.53, 0.02, 0.35))
cbar = plt.colorbar(Qr, cax=cax)
cbar.ax.tick_params(labelsize=18)
cbar.ax.set_title('g/kg',fontsize=22)
ax[0,0].set_title("298 K",fontsize=38)
ax[0,0].tick_params(labelsize=22)

ax[0,1].contourf(r300,z300,np.nanmean(Qr300,axis=0),cmap="PuBuGn",levels=np.arange(50,1200,100))
ax[0,1].contour(r300,z300,np.nanmean(Qs300,axis=0),cmap="cool",levels=(np.arange(500,2500,250)))
ax[0,1].contour(r300,z300,np.nanmean(Qg300,axis=0),colors="k",levels=(np.arange(50,250,50)))
ax[0,1].set_ylim(1,18)
ax[0,1].set_xlim(0,500)
ax[0,1].set_title("300 K",fontsize=38)
ax[0,1].set_yticks([])
ax[0,1].tick_params(labelsize=22)

ax[0,2].contourf(r302,z302,np.nanmean(Qr302,axis=0),cmap="PuBuGn",levels=np.arange(50,1200,100))
ax[0,2].contour(r302,z302,np.nanmean(Qs302,axis=0),cmap="cool",levels=(np.arange(500,2500,250)))
ax[0,2].contour(r302,z302,np.nanmean(Qg302,axis=0),colors="k",levels=(np.arange(50,250,50)))
ax[0,2].set_ylim(1,18)
ax[0,2].set_xlim(0,500)
ax[0,2].set_title("302 K",fontsize=38)
ax[0,2].set_yticks([])
ax[0,2].tick_params(labelsize=22)

ax[0,3].contourf(r304,z304,np.nanmean(Qr304,axis=0),cmap="PuBuGn",levels=np.arange(50,1200,100))
ax[0,3].contour(r304,z304,np.nanmean(Qs304,axis=0),cmap="cool",levels=(np.arange(500,2500,250)))
ax[0,3].contour(r304,z304,np.nanmean(Qg304,axis=0),colors="k",levels=(np.arange(50,250,50)))
ax[0,3].set_ylim(1,18)
ax[0,3].set_xlim(0,500)
ax[0,3].set_title("304 K",fontsize=38)
ax[0,3].set_yticks([])
ax[0,3].tick_params(labelsize=22)



Qc = ax[1,0].contourf(r298,z298,np.nanmean(Qc298,axis=0),cmap="YlOrRd",levels=np.arange(80,340,20))
ax[1,0].contour(r298,z298,np.nanmean(Qi298,axis=0),cmap="gist_heat",levels=np.arange(1,12,1))
ax[1,0].set_ylim(1,18)
ax[1,0].set_xlim(0,500)
ax[1,0].set_ylabel("Height [km]",fontsize=34)
cax = plt.axes((0.925, 0.11, 0.02, 0.35))
cbar = plt.colorbar(Qc, cax=cax)
cbar.ax.tick_params(labelsize=18)
cbar.ax.set_title('g/kg',fontsize=22)
ax[1,0].tick_params(labelsize=22)

ax[1,1].contourf(r300,z300,np.nanmean(Qc300,axis=0),cmap="YlOrRd",levels=np.arange(80,340,20))
ax[1,1].contour(r300,z300,np.nanmean(Qi300,axis=0),cmap="gist_heat",levels=np.arange(1,12,1))
ax[1,1].set_ylim(1,18)
ax[1,1].set_xlim(0,500)
ax[1,1].set_yticks([])
ax[1,1].tick_params(labelsize=22)

ax[1,2].contourf(r302,z302,np.nanmean(Qc302,axis=0),cmap="YlOrRd",levels=np.arange(80,340,20))
ax[1,2].contour(r302,z302,np.nanmean(Qi302,axis=0),cmap="gist_heat",levels=np.arange(1,12,1))
ax[1,2].set_ylim(1,18)
ax[1,2].set_xlim(0,500)
ax[1,2].set_yticks([])
ax[1,2].tick_params(labelsize=22)

ax[1,3].contourf(r304,z304,np.nanmean(Qc304,axis=0),cmap="YlOrRd",levels=np.arange(80,340,20))
ax[1,3].contour(r304,z304,np.nanmean(Qi304,axis=0),cmap="gist_heat",levels=np.arange(1,12,1))
ax[1,3].set_ylim(1,18)
ax[1,3].set_xlim(0,500)
ax[1,3].set_yticks([])
ax[1,3].tick_params(labelsize=22)


ax[0,0].text(30,17,"a",fontsize=28,weight="bold")
ax[0,1].text(30,17,"b",fontsize=28,weight="bold")
ax[0,2].text(30,17,"c",fontsize=28,weight="bold")
ax[0,3].text(30,17,"d",fontsize=28,weight="bold")

ax[1,0].text(30,17,"e",fontsize=28,weight="bold")
ax[1,1].text(30,17,"f",fontsize=28,weight="bold")
ax[1,2].text(30,17,"g",fontsize=28,weight="bold")
ax[1,3].text(30,17,"h",fontsize=28,weight="bold")

ax[1,3].text(-1000, -1.5, "Distance from TC Center [km]",fontsize=34)
```
