```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import xarray as xr
import numpy as np
from metpy.plots import ctables
```

```python
ncfile1 = xr.open_dataset("./data_files/fig2_vals_298K.nc")
refl_2km_298 = ncfile1.refl.values
CS_mask_298 = ncfile1.cs_breakdown.values
Storm_Mode_298 = ncfile1.modes.values
ncfile1.close()


#get inner and outer radii and storm center for plotting
df_298 = pd.read_csv("./data_files/TC_outer_inner_evolution_298K.csv")
time_298 = df_298.hours_after_init
r_inner_298 = df_298.rinner_v[191]
r_outer_298 = df_298.router_v[191]
x0_298 = int(df_298.x_center[191][1:4])
y0_298 = int(df_298.y_center[191][1:4])
```

```python
ncfile2 = xr.open_dataset("./data_files/fig2_vals_300K.nc")
refl_2km_300 = ncfile2.refl.values
CS_mask_300 = ncfile2.cs_breakdown.values
Storm_Mode_300 = ncfile2.modes.values
ncfile2.close()

#get inner and outer radii and storm center for plotting
df_300 = pd.read_csv("./data_files/TC_outer_inner_evolution_300K.csv")
time_300 = df_300.hours_after_init
r_inner_300 = df_300.rinner_v[191]
r_outer_300 = df_300.router_v[191]
x0_300 = int(df_300.x_center[191][1:4])
y0_300 = int(df_300.y_center[191][1:4])
```

```python
ncfile3 = xr.open_dataset("./data_files/fig2_vals_302K.nc")
refl_2km_302 = ncfile3.refl.values
CS_mask_302 = ncfile3.cs_breakdown.values
Storm_Mode_302 = ncfile3.modes.values
ncfile3.close()

#get inner and outer radii and storm center for plotting
df_302 = pd.read_csv("./data_files/TC_outer_inner_evolution_302K.csv")
time_302 = df_302.hours_after_init
r_inner_302 = df_302.rinner_v[191]
r_outer_302 = df_302.router_v[191]
x0_302 = int(df_302.x_center[191][1:4])
y0_302 = int(df_302.y_center[191][1:4])
```

```python
ncfile4 = xr.open_dataset("./data_files/fig2_vals_304K.nc")
refl_2km_304 = ncfile4.refl.values
CS_mask_304 = ncfile4.cs_breakdown.values
Storm_Mode_304 = ncfile4.modes.values
ncfile4.close()

#get inner and outer radii and storm center for plotting
df_304 = pd.read_csv("./data_files/TC_outer_inner_evolution_304K.csv")
time_304 = df_304.hours_after_init
r_inner_304 = df_304.rinner_v[191]
r_outer_304 = df_304.router_v[191]
x0_304 = int(df_304.x_center[191][1:4])
y0_304 = int(df_304.y_center[191][1:4])
```

```python
fig, ax = plt.subplots(3,4,figsize=(30,20))
ref_norm, ref_cmap = ctables.registry.get_with_steps('NWSStormClearReflectivity', -20, 0.5)

#horizontal reflectivity field at 2 km height
refl_plot = ax[0,0].pcolormesh(refl_2km_298,cmap=ref_cmap, norm=ref_norm, shading="auto")
ax[0,1].pcolormesh(refl_2km_300,cmap=ref_cmap, norm=ref_norm, shading="auto")
ax[0,2].pcolormesh(refl_2km_302,cmap=ref_cmap, norm=ref_norm, shading="auto")
ax[0,3].pcolormesh(refl_2km_304,cmap=ref_cmap, norm=ref_norm, shading="auto")

#circles showing r_inner and r_outer
ax[0,0].add_patch(plt.Circle((x0_298, y0_298), r_inner_298/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[0,0].add_patch(plt.Circle((x0_298, y0_298), r_outer_298/1.67, fill=False, edgecolor='k',linewidth=4))
ax[0,1].add_patch(plt.Circle((x0_300, y0_300), r_inner_300/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[0,1].add_patch(plt.Circle((x0_300, y0_300), r_outer_300/1.67, fill=False, edgecolor='k',linewidth=4))
ax[0,2].add_patch(plt.Circle((x0_302, y0_302), r_inner_302/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[0,2].add_patch(plt.Circle((x0_302, y0_302), r_outer_302/1.67, fill=False, edgecolor='k',linewidth=4))
ax[0,3].add_patch(plt.Circle((x0_304, y0_304), r_inner_304/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[0,3].add_patch(plt.Circle((x0_304, y0_304), r_outer_304/1.67, fill=False, edgecolor='k',linewidth=4))

ax[0,0].set_ylabel("Reflectivity at 2 km",fontsize=30)

#ax[0,1].set_yticks([])
#ax[0,2].set_yticks([])
#ax[0,3].set_yticks([])

#ax[0,0].set_xticks([])
#ax[0,1].set_xticks([])
#ax[0,2].set_xticks([])
#ax[0,3].set_xticks([])

ax[0,0].set_xlim(50,450)
ax[0,1].set_xlim(50,450)
ax[0,2].set_xlim(50,450)
ax[0,3].set_xlim(50,450)

ax[0,0].set_ylim(50,450)
ax[0,1].set_ylim(50,450)
ax[0,2].set_ylim(50,450)
ax[0,3].set_ylim(50,450)

ax[0, 0].set_aspect('equal', 'datalim')
ax[0, 1].set_aspect('equal', 'datalim')
ax[0, 2].set_aspect('equal', 'datalim')
ax[0, 3].set_aspect('equal', 'datalim')

ax[0,0].set_title("298 K",fontsize=40)
ax[0,1].set_title("300 K",fontsize=40)
ax[0,2].set_title("302 K",fontsize=40)
ax[0,3].set_title("304 K",fontsize=40)

ax[0,0].text(70,420,"a",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[0,1].text(70,420,"b",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[0,2].text(70,420,"c",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[0,3].text(70,420,"d",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")

cbar_ax1 = fig.add_axes([0.91, 0.658, 0.017, 0.222])
cbar1 = fig.colorbar(refl_plot, cax=cbar_ax1, fraction=0.04)
cbar1.set_label(label='dBZ',size=28,weight='bold')
cbar1.ax.tick_params(labelsize=18)


#convective/stratiform classification
mask_cmap = plt.cm.get_cmap('Paired', 7)
mask1 = ax[1,0].pcolormesh(CS_mask_298,cmap=mask_cmap,vmin=-0.9,vmax=5,shading="auto")
mask1.cmap.set_under('w')
mask2 = ax[1,1].pcolormesh(CS_mask_300,cmap=mask_cmap,vmin=-0.9,vmax=5,shading="auto")
mask2.cmap.set_under('w')
mask3 = ax[1,2].pcolormesh(CS_mask_302,cmap=mask_cmap,vmin=-0.9,vmax=5,shading="auto")
mask3.cmap.set_under('w')
mask4 = ax[1,3].pcolormesh(CS_mask_304,cmap=mask_cmap,vmin=-0.9,vmax=5,shading="auto")
mask4.cmap.set_under('w')

ax[1,0].add_patch(plt.Circle((x0_298, y0_298), r_inner_298/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[1,0].add_patch(plt.Circle((x0_298, y0_298), r_outer_298/1.67, fill=False, edgecolor='k',linewidth=4))
ax[1,1].add_patch(plt.Circle((x0_300, y0_300), r_inner_300/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[1,1].add_patch(plt.Circle((x0_300, y0_300), r_outer_300/1.67, fill=False, edgecolor='k',linewidth=4))
ax[1,2].add_patch(plt.Circle((x0_302, y0_302), r_inner_302/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[1,2].add_patch(plt.Circle((x0_302, y0_302), r_outer_302/1.67, fill=False, edgecolor='k',linewidth=4))
ax[1,3].add_patch(plt.Circle((x0_304, y0_304), r_inner_304/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[1,3].add_patch(plt.Circle((x0_304, y0_304), r_outer_304/1.67, fill=False, edgecolor='k',linewidth=4))

ax[1,0].set_ylabel("Precip. Classification",fontsize=30)

#ax[1,1].set_yticks([])
#ax[1,2].set_yticks([])
#ax[1,3].set_yticks([])

#ax[1,0].set_xticks([])
#ax[1,1].set_xticks([])
#ax[1,2].set_xticks([])
#ax[1,3].set_xticks([])

ax[1,0].set_xlim(50,450)
ax[1,1].set_xlim(50,450)
ax[1,2].set_xlim(50,450)
ax[1,3].set_xlim(50,450)

ax[1,0].set_ylim(50,450)
ax[1,1].set_ylim(50,450)
ax[1,2].set_ylim(50,450)
ax[1,3].set_ylim(50,450)

ax[1, 0].set_aspect('equal', 'datalim')
ax[1, 1].set_aspect('equal', 'datalim')
ax[1, 2].set_aspect('equal', 'datalim')
ax[1, 3].set_aspect('equal', 'datalim')

ax[1,0].text(70,420,"e",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[1,1].text(70,420,"f",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[1,2].text(70,420,"g",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[1,3].text(70,420,"h",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")

cbar_ax2 = fig.add_axes([0.91, 0.39, 0.017, 0.222])
cbar2 = fig.colorbar(mask1, 
                    cax=cbar_ax2, 
                    fraction=0.04, 
                    ticks=np.arange(0, 6, 1), 
                    boundaries=np.arange(-0.5, 6.5, 1),
                    orientation="vertical")
cbar2.ax.tick_params(labelsize=20)
cbar2.set_ticklabels(['Stra.', 'Conv.(1)', 'Conv.(2)', 'Conv.(3)', 'Conv.(4)', 'Conv.(5)'])


#Storm Mode classification
#cmap_selfdef_list = ['white','red', 'seagreen', 'dodgerblue', 'orange', 'purple']
cmap_selfdef_list = ['white',"#CC79A7", '#56B4E9', '#0072B2', "#F0E442", "#009E73"]
cmap_stormid = (mpl.colors.ListedColormap(cmap_selfdef_list))

stormid_298 = ax[2,0].pcolormesh(Storm_Mode_298, cmap=cmap_stormid, vmin=-0.1, vmax=5, shading="auto")
stormid_300 = ax[2,1].pcolormesh(Storm_Mode_300, cmap=cmap_stormid, vmin=-0.1, vmax=5, shading="auto")
stormid_302 = ax[2,2].pcolormesh(Storm_Mode_302, cmap=cmap_stormid, vmin=-0.1, vmax=5, shading="auto")
stormid_304 = ax[2,3].pcolormesh(Storm_Mode_304, cmap=cmap_stormid, vmin=-0.1, vmax=5, shading="auto")


ax[2,0].add_patch(plt.Circle((x0_298, y0_298), r_inner_298/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[2,0].add_patch(plt.Circle((x0_298, y0_298), r_outer_298/1.67, fill=False, edgecolor='k',linewidth=4))
ax[2,1].add_patch(plt.Circle((x0_300, y0_300), r_inner_300/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[2,1].add_patch(plt.Circle((x0_300, y0_300), r_outer_300/1.67, fill=False, edgecolor='k',linewidth=4))
ax[2,2].add_patch(plt.Circle((x0_302, y0_302), r_inner_302/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[2,2].add_patch(plt.Circle((x0_302, y0_302), r_outer_302/1.67, fill=False, edgecolor='k',linewidth=4))
ax[2,3].add_patch(plt.Circle((x0_304, y0_304), r_inner_304/1.67, fill=False, edgecolor='magenta',linewidth=4))
ax[2,3].add_patch(plt.Circle((x0_304, y0_304), r_outer_304/1.67, fill=False, edgecolor='k',linewidth=4))

ax[2,0].set_ylabel("Storm Modes",fontsize=30)

#ax[2,1].set_yticks([])
#ax[2,2].set_yticks([])
#ax[2,3].set_yticks([])

ax[2, 0].set_aspect('equal', 'datalim')
ax[2, 1].set_aspect('equal', 'datalim')
ax[2, 2].set_aspect('equal', 'datalim')
ax[2, 3].set_aspect('equal', 'datalim')

ax[2,0].set_xlim(50,450)
ax[2,1].set_xlim(50,450)
ax[2,2].set_xlim(50,450)
ax[2,3].set_xlim(50,450)

ax[2,0].set_ylim(50,450)
ax[2,1].set_ylim(50,450)
ax[2,2].set_ylim(50,450)
ax[2,3].set_ylim(50,450)


ax[2,0].text(70,420,"i",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[2,1].text(70,420,"j",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[2,2].text(70,420,"k",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")
ax[2,3].text(70,420,"l",fontsize=28,weight="bold",backgroundcolor="white",ha="center", va="center")

cbar_ax3 = fig.add_axes([0.91, 0.125, 0.015, 0.222])
cbar3 = fig.colorbar(stormid_302, 
                    cax=cbar_ax3, 
                    fraction=0.04, 
                    ticks=np.arange(1, 6, 1), 
                    boundaries=np.arange(0.5, 6.5, 1),
                    orientation="vertical")
cbar3.set_ticklabels(['DCC', 'OCC', 'WCC', 'DWCC', 'BSR'])
cbar3.ax.tick_params(labelsize=20)

fig.subplots_adjust(hspace=0.1,wspace=0.1)
```
